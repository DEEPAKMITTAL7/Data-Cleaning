```python
#loading important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('market_index_data_version1.csv') #loading the data
data
```

```python
df = pd.DataFrame(data)  # converting into the dataframe
df
```

```python
df.info() #checking the type of data
```

```python
df1 = df[(df['Index'] == 'NYA')] #creating a separte df for index=="NYA"
df1
```

```python
df1.info() #checking data type of the data
```

```python
df2 = df1.copy()  #creating a duplicate of df1
df2
```

```python
float_cols = df1.columns.drop(["Date", "Index"]) 
df2[float_cols] = df[float_cols].round().astype("Int64")  #converting float cols into int64
df2
```

```python
df2['Date'] = pd.to_datetime(df2['Date'],utc=True)  #coverting date into datetime format as per global standards
df2
```

```python
df2.info()
```

```python
df2.sort_values(by='Date')
```

```python
Year = df2['Date'].dt.year
```

```python
#plotting count plot by year
plt.figure(figsize=(20,10))
k = sns.countplot(data=df2, x=Year)
for i in k.containers:
    k.bar_label(i)
sns.set_style('darkgrid')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.show()
```
