```python
#loading important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('market_index_data_version1.csv') #loading the data
data
```

```python
df = pd.DataFrame(data)  # converting into the dataframe
df
```

```python
df.info() #checking the type of data
```

```python
df1 = df[(df['Index'] == 'NYA')] #creating a separte df for index=="NYA"
df1
```

```python
df1.info() #checking data type of the data
```

```python
df2 = df1.copy()  #creating a duplicate of df1
df2
```

```python
float_cols = df1.columns.drop(["Date", "Index"]) 
df2[float_cols] = df[float_cols].round().astype("Int64")  #converting float cols into int64
df2
```

```python
df2['Date'] = pd.to_datetime(df2['Date'],utc=True)  #coverting date into datetime format as per global standards
df2
```

```python
df2.info()
```

```python
df2.sort_values(by='Date')
```

```python
Year = df2['Date'].dt.year
```

```python
#plotting count plot by year
plt.figure(figsize=(20,10))
k = sns.countplot(data=df2, x=Year)
for i in k.containers:
    k.bar_label(i)
sns.set_style('darkgrid')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.show()
```

```python
#only one record found in 1965, dropping it

df3 = df2.drop(labels=0,axis=0)
df3
```

```python
#Plotting the target-date line graph

plt.figure(figsize=(12,8))
sns.lineplot(data=df3, x='Date', y='Adj Close')
sns.set_style('darkgrid')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.title('Adj Close by Date')
plt.grid()
plt.show()
```

```python
#let's drop the row 289 with all NA
df4 = df3.dropna(thresh=6)
df4[df4.isna().any(axis=1)]
```

```python
df5 = df4.fillna(method='pad')
df5[df5.isna().any(axis=1)]
```

```python
# Index = 831
df5.drop(df5[(df5['Adj Close']>5000)&(df5.Date.dt.year<1975)].index, inplace=True)
```

```python
# Index = 852
df5.drop(df5[(df5['Adj Close']<1000)&(df5.Open>5000)].index, inplace=True) 
```

```python
# Index = 829, 833
df5.drop(df5[(df5['Adj Close']<1000)&(df5.High>5000)].index, inplace=True)
```

```python
# Dropping the "Volume" column because of the high missing values
df6 = df5.drop("Volume", axis= 1)
```

```python
df6.describe()
```

```python
# Adj Close vs. Date
plt.figure(figsize = (20, 10))
sns.lineplot(data=df6, x="Date", y="Adj Close")
sns.set_style('darkgrid')
plt.title("Adj Close by Date")
plt.ylabel("Adj Close", fontsize=11)
plt.xlabel("Date", fontsize=11)
plt.show()
```
